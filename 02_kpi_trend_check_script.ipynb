```python
import pandas as pd
```

```python
df = pd.read_csv('monthly retention and churn rate.csv')
```

```python
# active users at start
df.plot(x='month', y='active_at_start')
```

Notes: number of active users increases consistently. 

```python
df['active_at_start'].describe()
```

```python
# monthly churn rate
df.plot(x='month', y='churn_rate_percentage')
```

Notes: the churn rate increases dramatically at the begining, at local maximum at the end of the first year (2023), then drops the its lowest at end of second year (2024), then starts to increase again at the third year (2025). 

```python
df['churn_rate_percentage'].describe() 
```
